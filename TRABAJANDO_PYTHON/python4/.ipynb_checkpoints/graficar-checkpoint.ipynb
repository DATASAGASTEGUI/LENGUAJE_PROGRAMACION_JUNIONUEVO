<span style="color:blue; font-size:15px">EJERCICIO 1</span><br>
Graficar la función trigonométrica seno:

$\begin{equation}
y = f(x) = \sin(x)
\end{equation}$

6000 MHz = 6000 000 000 Hz = ciclos/segundo
1Hz = 1 ciclo/segundo

2 lecturas por cada ciclo = 12 000 000 000 bits


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,2*np.pi,360)
y = np.sin(x)

plt.plot(x,y,label="SIN")
plt.title("FUNCION TRIGONOMETRIA SENO")
plt.xlabel("X")
plt.ylabel("Y = SIN(X)")
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

frecuencia = 3

x = np.linspace(0,2*np.pi,360)
y = np.sin(frecuencia * x)

plt.plot(x,y,label="SIN")
plt.title("FUNCION TRIGONOMETRIA SENO")
plt.xlabel("X")
plt.ylabel("Y = SIN(X)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

meses = ["ENERO","FEBRERO","MARZO"] # EJE X
ventas = [45384, 23434, 13434] # EJE Y

plt.figure(figsize=(12,6))
plt.bar(meses,ventas,color='lightgreen')
plt.title("VENTAS POR MES")
plt.xlabel("MES")
plt.ylabel("VENTAS")
plt.show()
```
